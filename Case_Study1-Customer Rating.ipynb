```python
import pandas as pd
```

```python
rating = pd.Series([5, 4, 3, 5, 2, 1, 4, 5, 3, 4, 2, 5, 5, 3, 1, 4, 4, 5, 2, 3],name='Rating')
```

### Basic Info
Print the first 5 ratings.

```python
rating.head()
```

```python
rating.size
```

```python
rating.unique()
```

```python
rating.value_counts()
```

### 2. Descriptive Statistics
Calculate mean, median, mode of the ratings.

Find the 25th, 50th, and 75th percentiles.

```python
rating.mean()
```

```python
rating.median()
```

```python
rating.mode()[0]
```

```python
rating.describe()
```

```python
rating.quantile(0.25)
```

```python
rating.quantile(0.5)
```

### 3. Filtering Data
Show all ratings equal to 5.

Show ratings between 2 and 4 (inclusive).

Count how many ratings are below 3.

```python
rating[rating == 5]
```

```python
rating[rating.between(2,4)]
```

```python
rating[rating <3].size
```

### 4. Transformations
Replace all 1s with 0 (assuming 1-star = missing/bad data).

Map ratings to text:
{5: 'Excellent', 4: 'Good', 3: 'Average', 2: 'Poor', 1: 'Very Poor'}

```python
rating.replace(1,0,inplace=True)
```

```python
rating
```

```python
rating.replace({5:'Excellent',4:'Good',3:'Average',2:'Bad',0:'Very Poor'})
```
