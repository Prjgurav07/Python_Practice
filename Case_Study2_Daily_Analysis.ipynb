### Data Cleaning
Replace 0 values with the mean or median sales.

Create a new Series where sales are categorized:
>= 2100 → "High"
1800–2099 → "Medium"
< 1800 → "Low"

```python
sales.replace(0,sales.mean())
```

```python
def sales_cat(value):
    if value>=2100:
        return 'High'
    elif value >=1800 and value <= 2099:
        return 'Medium'
    else:
        return 'Low'

sales1 = sales.apply(sales_cat)
sales1
```
